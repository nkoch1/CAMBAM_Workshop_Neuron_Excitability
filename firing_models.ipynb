```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# LIF model
dt = 1
tmax = 2e3
T = np.arange(0,tmax,dt)
N = len(T)


gL = 0.1
dx = lambda x,I: -gL*(x+60) + I

x = -60*np.ones([N,1])
for i in range(0,N-1):
    if(x[i]>-30):
        x[i] = 60
        x[i+1] = -60
    else:
        x[i+1] = x[i] + dt*dx(x[i],2*np.random.randn()+2)
plt.plot(T,x)
```

```python
# QIF model
dt = 1
tmax = 2e3
T = np.arange(0,tmax,dt)
N = len(T)


gL = 1e-4
dx = lambda x,I: gL*x**2 + I

x = -60*np.ones([N,1])
for i in range(0,N-1):
    if(x[i]>-30):
        x[i] = 60
        x[i+1] = -60
    else:
        x[i+1] = x[i] + dt*dx(x[i],-0.1+2*np.random.randn())
plt.plot(T,x)
```

```python
# Izhikevich model
dt = 0.5
tmax = 2e3
T = np.arange(0,tmax,dt)
N = len(T)

a = 0.02
b = 0.25
c = -65
d = 2

dv = lambda v,u,I: 0.04*v**2+5*v+140-u+I
du = lambda v,u: a*(b*v-u)

v = c*np.ones([N,1])
u = b*v
for i in range(0,N-1):
    if(v[i]>=30):
        v[i] = 30
        v[i+1] = c
        u[i+1] = u[i]+d
    else:
        v[i+1] = v[i] + dt*dv(v[i],u[i],np.random.randn())
        u[i+1] = u[i] + dt*du(v[i],u[i])
plt.plot(T,v)
```
